```python
import warnings
warnings.filterwarnings('ignore')
#Giờ load vào Python — tự viết code kết nối + EDA tổng quan:
##1. Kết nối MySQL → bank_churn_db đổi thành marketing_db
import mysql.connector
import pandas as pd
conn = mysql.connector.connect(
    host='localhost',
    user='root',
    password='Phuongthu@1001',
    database='marketing_db'
)
##2. Load bankchurners_clean đổi thành ml_project1_data
df = pd.read_sql("SELECT * FROM ml_project1_data", conn)
print(df.shape)
print(df.dtypes)
print(df['Response'].value_counts())
##3. Drop Z_CostContact, Z_Revenue
df.drop(columns=['Z_CostContact', 'Z_Revenue'], inplace=True)
##4. df.describe() cho các cột Income, Year_Birth, Recency
df.describe()
df[['Income','Year_Birth','Recency']].describe()
##=> 2216 dòng, 29 cột . Response: 1883 vs 333 (15% positive) → imbalanced ~5.6:1 (có sự mất cân bằng lớp, ỷ lệ giữa hai lớp xấp xỉ 5.6:1.)
# Dùng tự động tra cột bị lỗi
# Định nghĩa hàm trinh sát tự động
def check_data_sanity(dataframe):
    import pandas as pd
    num_cols = dataframe.select_dtypes(include=['int64', 'float64']).columns
    sanity_report = []
    for col in num_cols:
        Q1 = dataframe[col].quantile(0.25)
        Q3 = dataframe[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        num_outliers = len(dataframe[(dataframe[col] < lower_bound) | (dataframe[col] > upper_bound)])
        skew_val = dataframe[col].skew()
        
        sanity_report.append({
            'Tên Cột': col,
            'Số dòng Outlier': num_outliers,
            'Tỷ lệ Outlier (%)': round((num_outliers / len(dataframe)) * 100, 2),
            'Độ lệch (Skewness)': round(skew_val, 4)
        })
    report_df = pd.DataFrame(sanity_report)
    # Chỉ trả về những cột thực sự có bất thường để đỡ rối mắt
    return report_df[(report_df['Số dòng Outlier'] > 0) | (abs(report_df['Độ lệch (Skewness)']) > 1)]

print(" BÁO CÁO TRINH SÁT LỖI TỰ ĐỘNG ")
print(check_data_sanity(df))
#
##| Skewness                   | Đánh giá       | Nhận xét                   |   
#| -------------------------- | -------------- | -------------------------- |
#| -0.5 đến 0.5               | Gần đối xứng   | Dữ liệu khá cân đối        |
#| 0.5 đến 1 hoặc -1 đến -0.5 | Lệch nhẹ       | Có xu hướng lệch trái/phải |
#| > 1                        | Lệch phải mạnh | Có một số giá trị rất lớn  |
#| < -1                       | Lệch trái mạnh | Có một số giá trị rất nhỏ  |
##Ra quyết định xử lý
#Tình huống	Quyết định
#Tỷ lệ thấp, giá trị vô lý->	Xem xét loại bỏ
#Tỷ lệ thấp nhưng hợp lý->	Giữ lại
#Tỷ lệ cao do phân phối lệch	->Giữ lại hoặc biến đổi
#Biến chi tiêu/thu nhập có giá trị rất lớn	->Thường giữ lại vì có ý nghĩa
#Biến nhị phân (0/1)	->Bỏ qua kiểm tra IQR
##Tỷ lệ Outlier 	Mức độ	       Ý nghĩa                   	Hướng xử lý
#< 1%	        Rất thấp	Chỉ xuất hiện một vài giá trị khác biệt	Kiểm tra từng trường hợp, có thể loại bỏ nếu là lỗi dữ liệu
#1% – 5%	Thấp đến trung bình	   Có sự khác biệt giữa một số nhóm quan sát	Kết hợp xem Skewness và ý nghĩa biến trước khi quyết định
#5% – 10%	Cao	Dữ liệu có thể phân tán lớn hoặc phân phối lệch	Không xóa ngay, kiểm tra phân phối và đặc điểm dữ liệu
#> 10%	Rất cao
#
#Income: Income=666,666 với mean=52,247 → outlier rõ ràng, cần kiểm tra IQR.
#Year_Birth=1893: Năm 2014 - 1893 = 121 tuổi — không thực tế cho khách hàng marketing campaign.
##Chạy kiểm tra outliers bằng IQR — tự viết theo logic đã học ở tài liệu trước:
#1. Tính Q1, Q3, IQR cho Income và Year_Birth
# Check Null trước khi tính IQR = Q3(75%)- Q1(25%) ( tránh bẫy mising _dữ liệu trống )
print(" MISSING VALUES ")
print(df[['Income', 'Year_Birth']].isnull().sum())

# 2. Tính IQR cho Income
Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)
IQR = Q3 - Q1
lower_income = Q1 - 1.5 * IQR
upper_income = Q3 + 1.5 * IQR

print("INCOME OUTLIER BOUNDS ")
print(f"Lower: {lower_income:.0f} | Upper: {upper_income:.0f}") # chữ f trước"" là ép dạng trong ngoặc thành dạng string nếu bỏ f thì sdx là chứa văn bản bth k chứa biến
income_outliers = df[(df['Income'] < lower_income) | (df['Income'] > upper_income)]
print(f"Số dòng outlier: {len(income_outliers)}") # len(): Chèn kết quả đếm vào chuỗi f-string
print(income_outliers[['ID', 'Income', 'Year_Birth']])

# 3. Tính IQR cho Year_Birth
Q1_yb = df['Year_Birth'].quantile(0.25)
Q3_yb = df['Year_Birth'].quantile(0.75)
IQR_yb = Q3_yb - Q1_yb
lower_yb = Q1_yb - 1.5 * IQR_yb
upper_yb = Q3_yb + 1.5 * IQR_yb

print("YEAR_BIRTH OUTLIER BOUNDS")
print(f"Lower: {lower_yb:.0f} | Upper: {upper_yb:.0f}")

yb_outliers = df[(df['Year_Birth'] < lower_yb) | (df['Year_Birth'] > upper_yb)]
print(f"Số dòng outlier: {len(yb_outliers)}")
print(yb_outliers[['ID', 'Income', 'Year_Birth']])
#NHẬN XÉT:
#-Income outliers (8 dòng):7 dòng có Income 153k-162k → cao nhưng có thể thật (người giàu)
                        #1 dòng Income=666,666 → rõ ràng là lỗi nhập liệu
#-Year_Birth outliers (3 dòng):1900, 1893, 1899 → tuổi 115-121 → không thể thật
#=>7 dòng Income cao — cần xác nhận thêm bằng cách nhìn vào các cột khác (chi tiêu, số lần mua) — người thu nhập cao thật sự sẽ có spending cao tương ứng.
# Chạy để ktra:

# BƯỚC KHẢO SÁT HÀNH VI ĐỂ QUYẾT ĐỊNH GIỮ HAY XÓA 7 DÒNG INCOME CAO

# 1. Lọc ra 7 dòng có thu nhập cao (nằm trong khoảng 153k đến 162k, loại trừ hẳn ông cụ 666666)
high_income = df[(df['Income'] > upper_income) & (df['Income'] < 666666)]

print(" PROFILE CHI TIÊU CỦA 7 ĐẠI GIA ")
# In ra các cột hành vi mua sắm xem họ tiêu tiền thế nào
print(high_income[['ID', 'Income', 'MntWines', 'MntMeatProducts', 'MntFruits', 'NumStorePurchases', 'Year_Birth']])

print("\ MỨC TIÊU DÙNG TRUNG BÌNH CỦA TOÀN SIÊU THỊ ")
# Tính trung bình toàn bộ dataset để làm bàn cân đối chiếu
print(df[['Income', 'MntWines', 'MntMeatProducts', 'MntFruits', 'NumStorePurchases']].mean().round(0))
#Nếu thu nhập của họ cao gấp 3 lần người bình thường, nhưng số tiền họ mua rượu (MntWines), mua thịt (MntMeatProducts) cũng cao vượt trội, thì họ là Người giàu thật (Giữ lại để phân tích nhóm VIP). Ngược lại,
# Xem profile đầy đủ của 7 dòng Income cao (trừ 666666)
# Xem profile đầy đủ của 7 dòng Income cao (trừ 666666)
high_income = df[(df['Income'] > upper_income) & (df['Income'] < 666666)]
print(high_income[['ID', 'Income', 'MntWines', 'MntMeatProducts', 
                    'MntFruits', 'NumStorePurchases', 'Year_Birth']])

# So sánh với trung bình toàn dataset
print("\n=== TRUNG BÌNH TOÀN DATASET ===")
print(df[['Income', 'MntWines', 'MntMeatProducts', 
          'MntFruits', 'NumStorePurchases']].mean().round(0))
#=> sO Sánh với tb, ta thấy:
# Thu nhập cao hơn tb; trong khi chi tiêu lại thấp hơn: MntWWitnes( thấp hơn tb), MntProducts( lộn xộn có cao có thấp), NnFriurt+ nUM (THÂP HƠN tb)
#=>7 dòng này có Income cao nhưng spending pattern không nhất quán — không thể xác nhận là thật hay lỗi. Xử lý bằng Winsorization — giới hạn về percentile 99 thay vì xóa, để không mất data.
#QUYẾT ĐỊNH XỬ LÍ OUTLINE:
#Income = 666,666  → XÓA (lỗi nhập liệu rõ ràng)
#Income 153k-162k  → Winsorization về percentile 99
#Year_Birth 1893-1900 → XÓA (không thực tế)
# BƯỚC 4: TIẾN HÀNH DỌN DẸP DATA THEO QUYẾT ĐỊNH XỬ LÝ OUTLIERS

# --- 1. XÓA BỎ CÁC DÒNG QUÁ CỰC ĐOAN (Income = 666,666 và Người sinh trước năm 1900) ---
# Logic: Giữ lại những người có Thu nhập khác 666666 VÀ có Năm sinh từ 1900 trở đi
df_clean = df[(df['Income'] != 666666) & (df['Year_Birth'] > 1900)]
# --- 2. ÁP DỤNG WINSORIZATION CHO CỘT INCOME (Đưa nhóm 153k-162k về Percentile 99) --- #ép 
# Bước A: Tìm giá trị cắt tại Percentile 99 của tập dữ liệu đã lọc bớt rác
p99_income = df_clean['Income'].quantile(0.99)
print(f"Giá trị tại Percentile 99 của Income là: {p99_income:.0f}")

# Bước B: Ép trần (Cap) giá trị. Những ai có Income vượt quá p99 sẽ bị kéo về đúng mức p99
# Hàm np.where(điều_kiện, giá_trị_nếu_đúng, giá_trị_nếu_sai) của numpy làm việc này rất mượt
import numpy as np
df_clean['Income'] = np.where(df_clean['Income'] > p99_income, p99_income, df_clean['Income'])
#np.where để Winsorization? Hàm này hoạt động giống như lệnh IF trong Excel. Nó quét qua toàn bộ cột Income, nếu gặp ông nào $> p99$ (
# --- 3. KIỂM TRA LẠI KẾT QUẢ SAU KHI XỬ LÝ ---
print("\n THỐNG KÊ SAU KHI LÀM SẠCH ")
print(f"Kích thước dữ liệu gốc: {df.shape}")
print(f"Kích thước dữ liệu sạch (df_clean): {df_clean.shape}")

# In lại giá trị max để chắc chắn đại gia 666k đã bay màu và nhóm 160k đã bị ép về p99
print(f"Thu nhập lớn nhất hiện tại (Max Income): {df_clean['Income'].max():.0f}")
#df_clean['Income'].max() bắt buộc phải trả ra đúng bằng con số p99_income mà bạn vừa tính ở trên. Nếu hai số đó bằng nhau, bạn đã Winsorization thành công mỹ mãn!
print(f"Năm sinh nhỏ nhất hiện tại (Min Year_Birth): {df_clean['Year_Birth'].min()}")
#Bước 1.5 gần xong! Còn 1 việc — cột Dt_Customer đang là text, cần convert sang datetime để tính Customer_Tenure (số ngày gắn bó).
# =====================================================================

# 1. Convert cột Dt_Customer từ dạng Object (Text) sang dạng Datetime
# Thêm errors='coerce' để nếu có ô nào lỗi định dạng thì nó tự chuyển thành NaT (Null) thay vì báo lỗi dừng code
df_clean['Dt_Customer'] = pd.to_datetime(df_clean['Dt_Customer'], errors='coerce')

# 2. Tạo cột Age (Tuổi) dựa trên mốc năm 2024 như bạn vạch ra
df_clean['Age'] = 2024 - df_clean['Year_Birth']

# 3. Tạo cột Customer_Tenure (Số ngày khách hàng đã gắn bó với siêu thị)
# Bước A: Xác định "ngày collect data" tự động bằng cách lấy ngày muộn nhất trong cột dữ liệu
max_date = df_clean['Dt_Customer'].max()
print(f"Ngày tương tác muộn nhất trong dữ liệu: {max_date.strftime('%Y-%m-%d')}")

# Bước B: Tính số ngày gắn bó bằng cách lấy (Ngày thu thập - Ngày đăng ký)
# Dùng .dt.days để ép kiểu kết quả từ dạng chuỗi thời gian (Timedelta) về dạng số nguyên (Integer)
df_clean['Customer_Tenure'] = (max_date - df_clean['Dt_Customer']).dt.days

# =====================================================================
# KIỂM TRA LẠI KẾT QUẢ SAU KHI TẠO BIẾN
print("\n=== THÔNG TIN DỮ LIỆU SAU KHI BIẾN ĐỔI ===")
print(df_clean[['Dt_Customer', 'Year_Birth', 'Age', 'Customer_Tenure']].head())

# Kiểm tra lại kiểu dữ liệu (dtypes) xem Dt_Customer đã thành datetime64 chưa, Age và Tenure có phải là số không
print("\nKiểu dữ liệu của các cột mới:")
print(df_clean[['Dt_Customer', 'Age', 'Customer_Tenure']].dtypes)
# =====================================================================
# BƯỚC 5: XỬ LÝ DỮ LIỆU KHUYẾT THIẾU (MISSING VALUES) CHO CỘT INCOME

# 1. Kiểm tra xem hiện tại có chính xác bao nhiêu dòng bị khuyết Income
missing_before = df_clean['Income'].isnull().sum()
print(f"Số dòng bị thiếu Income trước khi xử lý: {missing_before}")

# 2. Áp dụng kỹ thuật điền khuyết nâng cao: Điền bằng Median theo nhóm Trình độ học vấn (Education)
# Dùng transform('median') để Pandas tự động tính và map giá trị tương ứng vào từng dòng
df_clean['Income'] = df_clean['Income'].fillna(df_clean.groupby('Education')['Income'].transform('median'))

# 3. Kiểm tra lại kết quả xem còn dòng nào bị khuyết không
missing_after = df_clean['Income'].isnull().sum()
print(f"Số dòng bị thiếu Income sau khi xử lý: {missing_after}")

# In thử thống kê mô tả lại của cột Income để chắc chắn dữ liệu đã đầy đủ và không bị biến động cấu trúc
print("\n=== THỐNG KÊ CỘT INCOME SAU KHI ĐIỀN KHUYẾT ===")
print(df_clean['Income'].describe())
#mean = 51724 và 50% (Median) = 51371: Đây chính là "chứng chỉ vàng" chứng minh bạn đã điền khuyết cực kỳ thành công. Khi giá trị trung bình (mean) và trung vị (median) sát rạt nhau như thế này (chênh lệch chưa tới 400 đơn vị), điều đó có nghĩa là phân phối của cột Income đã hoàn toàn bình ổn, không còn bị lệch nặng (Skewed) hay bị méo mó bởi các giá trị dị biệt nữa.
##Hiện tại, dữ liệu nhân khẩu học và thời gian đã sạch sẽ. Nhưng nếu bạn lướt qua các cột còn lại trong dataset, bạn sẽ thấy siêu thị đang lưu trữ dữ liệu chi tiêu rất chi tiết và lặt vặt: MntWines (Rượu), MntFruits (Trái cây), MntMeatProducts (Thịt), MntFishProducts (Cá), MntSweetProducts (Đồ ngọt), MntGoldProds (Vàng/Đồ xa xỉ).
#Trong thực tế làm dự án Marketing (đặc biệt là bài toán phân khúc khách hàng), nếu để quá nhiều cột nhỏ lẻ như vậy, mô hình sẽ bị loãng. Sếp hoặc khách hàng chỉ muốn biết một câu hỏi cốt lõi: "Ông này tổng cộng đã cúng cho siêu thị bao nhiêu tiền và mua bao nhiêu lần?"
#Vì vậy, ý tưởng tiếp theo là chúng ta sẽ gom nhóm để tạo ra 2 biến "Tổng lực" đại diện cho giá trị của khách hàng:
# =====================================================================
# BƯỚC 6: TẠO CÁC FEATURE TỔNG LỰC ĐỂ PHỤC VỤ PHÂN TÍCH (COMPREHENSIVE FEATURES)
# =====================================================================

# 1. Tạo cột Total_Spent: Tổng số tiền khách hàng đã chi tiêu cho tất cả các mặt hàng
df_clean['Total_Spent'] = (
    df_clean['MntWines'] + 
    df_clean['MntFruits'] + 
    df_clean['MntMeatProducts'] + 
    df_clean['MntFishProducts'] + 
    df_clean['MntSweetProducts'] + 
    df_clean['MntGoldProds']
)

# 2. Tạo cột Total_Purchases: Tổng số lần mua hàng qua các kênh khác nhau
# (Loại trừ cột NumWebVisitsPerMonth vì đó là số lần ghé thăm web, không phải mua thật)
df_clean['Total_Purchases'] = (
    df_clean['NumWebPurchases'] + 
    df_clean['NumCatalogPurchases'] + 
    df_clean['NumStorePurchases']
)

# =====================================================================
# KIỂM TRA LẠI KẾT QUẢ VÀ ĐỌC SƠ BỘ INSIGHT

print("=== THỐNG KÊ CHI TIÊU VÀ SỐ LẦN MUA SẮM TỔNG HỢP ===")
print(df_clean[['Total_Spent', 'Total_Purchases']].describe())

# In thử 5 dòng đầu để check xem toán tử cộng có khớp logic không
print("\nCheck thử 5 dòng đầu tiên:")
print(df_clean[['ID', 'Income', 'Total_Spent', 'Total_Purchases']].head())
#=> Đọc vị" Insight từ 2 Feature tổng lực bạn vừa tạo
#1. Về chi tiêu (Total_Spent):
#- Khách hàng tiết kiệm nhất: Chỉ tiêu vỏn vẹn 5$.
#- Đại gia chịu chi nhất (Max): Tiêu tới 2,525$! (Nhóm VIP này chính là lý do vì sao bài trước ta dùng Winsorization để giữ họ lại chứ không xóa thẳng tay, họ mang lại nguồn thu lớn cho siêu thị).
#- Con số giật mình (Median vs Mean): Giá trị trung vị là 397$ nhưng trung bình lên tới 607$. Điều này chứng tỏ dữ liệu chi tiêu bị lệch phải nặng (Skewed) — một nhóm nhỏ khách hàng giàu đang "gánh" doanh thu cho cả siêu thị.
#2.Về số lần mua sắm (Total_Purchases):
#- Trung bình một khách hàng mua khoảng 12-13 lần qua các kênh.
#- Người mua nhiều nhất lên tới 32 lần.
#- Ngược lại, có người mua 0 lần (min = 0) nhưng hãy nhìn lên Total_Spent min lại là 5$.
#=> Bẫy logic xuất hiện: Tại sao mua 0 lần mà vẫn tốn 5$? À, có thể họ mua qua một kênh khác không nằm trong 3 cột ta cộng (ví dụ: mua qua deal khuyến mãi NumDealsPurchases), hoặc đây là phí mở thẻ. Thấy chưa, đào sâu data thực tế nó hay ở chỗ đó!

#CÂU HỎI KINH DOANH 1: NHÓM THẾ HỆ NÀO (AGE GROUP) CHỊU CHI NHẤT
#Để làm việc này, chúng ta sẽ chia cột Age thành 4 nhóm tuổi bằng hàm pd.cut của Pandas:

# BƯỚC 7: PHÂN NHÓM TUỔI VÀ PHÂN TÍCH MỨC ĐỘ ĐÓNG GÓP DOANH THU
# =====================================================================

# 1. Định nghĩa các khoảng cắt (bins) và nhãn (labels) cho nhóm tuổi
bins = [0, 30, 45, 60, 150]
labels = ['Under 30', '30-45', '45-60', 'Above 60']

# 2. Tạo cột Age_Group bằng hàm pd.cut
df_clean['Age_Group'] = pd.cut(df_clean['Age'], bins=bins, labels=labels)

# 3. Gom nhóm theo Age_Group để tính: Trung bình Thu nhập, Tổng chi tiêu, và Số lượng khách mỗi nhóm
age_analysis = df_clean.groupby('Age_Group', observed=False).agg({
    'ID': 'count',                       # Đếm số lượng khách hàng
    'Income': 'mean',                    # Thu nhập trung bình của nhóm
    'Total_Spent': 'sum',                # Tổng số tiền nhóm này đóng góp (Doanh thu)
    'Total_Purchases': 'mean'            # Số lần mua hàng trung bình
}).rename(columns={'ID': 'Customer_Count', 'Total_Spent': 'Total_Revenue'})

# 4. Tính thêm tỷ lệ % đóng góp doanh thu của từng nhóm để sếp dễ nhìn
total_rev = age_analysis['Total_Revenue'].sum()
age_analysis['Revenue_Share (%)'] = round((age_analysis['Total_Revenue'] / total_rev) * 100, 2)

print("=== BẢNG PHÂN TÍCH KHÁCH HÀNG THEO NHÓM TUỔI ===")
print(age_analysis.round(0))

#📊 Đọc Vị Insight Từ Bảng Thống Kê Nhóm Tuổi
#- 💥 Nhóm "Quyền lực ngầm" (45-60 và Trên 60 tuổi): Hãy nhìn vào cột Revenue_Share (%). Hai nhóm này cộng lại gánh tới 80% tổng doanh thu của siêu thị ($41\% + 39\%$). Thu nhập (Income) của họ cũng ở mức cao nhất ($50k - $57k).
#- 💡 Business Action: Chiến dịch Marketing tiếp theo phải tập trung đánh mạnh vào nhóm này. Các sản phẩm bổ dưỡng, rượu vang cao cấp, hoặc dịch vụ chăm sóc khách hàng VIP nên được ưu tiên thiết kế riêng cho họ.
#📉 Nhóm Trẻ (Under 30) — Thị trường ngách hay "Dead segment"?: Nhóm này chỉ có vỏn vẹn 10 người và đóng góp đúng 1% doanh thu. Nhưng hãy nhìn vào Total_Purchases (14.0) và Income ($58,181) ->Họ rất giàu và siêng mua sắm, chỉ là siêu thị chưa biết cách thu hút số lượng lớn tệp khách hàng trẻ này mà thôi.

##CÂU HỎI KINH DOANH 2(KÊNH BÁN HÀNG): KÊNH MUA SẮM NÀO MANG LẠI HIỆU QUẢ NHẤT CHO SIÊU THỊ
#Siêu thị đang phân phối qua 3 kênh chính:
# =====================================================================
# BƯỚC 8: PHÂN TÍCH HIỆU QUẢ CÁC KÊNH BÁN HÀNG (CHANNEL ANALYSIS)
# =====================================================================

# 1. Tính tổng số đơn hàng của từng kênh trên toàn bộ hệ thống
total_web = df_clean['NumWebPurchases'].sum() #NumWebPurchases: Mua trực tuyến qua Website.
total_catalog = df_clean['NumCatalogPurchases'].sum() #NumCatalogPurchases: Mua qua sách hướng dẫn (Catalog) gửi tận nhà.
total_store = df_clean['NumStorePurchases'].sum() #NumStorePurchases: Đến mua trực tiếp tại Cửa hàng.

# Tổng tất cả các đơn hàng để làm mẫu số tính tỷ lệ %
all_orders = total_web + total_catalog + total_store

# 2. Đóng gói kết quả thành một DataFrame nhỏ để quan sát
channel_report = pd.DataFrame({
    'Kênh Bán Hàng': ['Website', 'Catalog', 'Cửa hàng trực tiếp'],
    'Tổng Số Đơn Hàng': [total_web, total_catalog, total_store],
    'Tỷ Lệ Đóng Góp (%)': [
        round((total_web / all_orders) * 100, 1),
        round((total_catalog / all_orders) * 100, 1),
        round((total_store / all_orders) * 100, 1)
    ]
})

print("=== BÁO CÁO HIỆU QUẢ KÊNH PHÂN PHỐI ===")
print(channel_report.to_string(index=False))
#=> 📊 "Đọc vị" Insight từ Báo cáo Kênh Phân Phối
#- 🏬 Cửa hàng trực tiếp (Store) là "Gà đẻ trứng vàng": Chiếm tới 46.2% tổng số đơn hàng (gần một nửa doanh số). Điều này hoàn toàn đồng nhất với Insight ở bước trước: Tệp khách hàng gánh doanh thu của siêu thị là nhóm Trung niên và Lớn tuổi (trên 45 tuổi). Nhóm này có thói quen thích ra tận nơi xem hàng, đẩy xe và quẹt thẻ trực tiếp.
#-🌐 Website (Digital Channel) không hề kém cạnh: Đứng thứ hai với 32.5% đơn hàng. Đây là một con số rất lớn cho thấy kênh Online đang hoạt động cực kỳ hiệu quả và có tiềm năng tăng trưởng mạnh nếu siêu thị biết cách tối ưu hình thức chạy quảng cáo số.
#📬 Catalog (Gửi bộ tài liệu tận nhà): Tuy là phương thức cổ điển nhất nhưng vẫn mang lại 21.3% đơn hàng. Ở các nước phát triển, việc lật giở catalog giấy và gọi điện đặt hàng vẫn là sở thích của các cụ già có điều kiện tài chính cao.

#CÂU HỎI KINH DOANH 3 (Khảo Sát Chân Dung Nhóm 15% Khách Hàng VIP): "Ai là người thực sự phản hồi lại quảng cáo của chúng ta (Response = 1_ ĐỒNG Ý)?"
#Chúng ta sẽ bóc tách chân dung của hai nhóm này ra để so sánh xem nhóm đồng ý mua (Response = 1) có điểm gì khác biệt vượt trội về: Thu nhập (Income), Tuổi tác (Age), Tổng chi tiêu (Total_Spent) và Số ngày gắn bó (Customer_Tenure).
# =====================================================================
# BƯỚC 9: ĐỘT PHÁ INSIGHT - PHÂN TÍCH CHÂN DUNG KHÁCH HÀNG PHẢN HỒI (RESPONSE)
# =====================================================================

# Gom nhóm theo cột Response để tính toán các giá trị trung bình đặc trưng
target_analysis = df_clean.groupby('Response').agg({
    'ID': 'count',                        # Số lượng khách mỗi nhóm
    'Age': 'mean',                        # Tuổi trung bình
    'Income': 'mean',                     # Thu nhập trung bình
    'Total_Spent': 'mean',                # Mức chi tiêu trung bình
    'Customer_Tenure': 'mean'             # Số ngày gắn bó trung bình
}).rename(columns={'ID': 'Số lượng khách hàng'})

# Đổi tên chỉ mục (Index) từ 0/1 thành Từ chối/Đồng ý để báo cáo nhìn rõ ràng
target_analysis.index = ['Từ chối (Response=0)', 'Đồng ý (Response=1)']

print("=== BẢNG ĐỐI CHIẾU CHÂN DUNG KHÁCH HÀNG ===")
print(target_analysis.round(1))
#=>👁️ "Đọc vị" Chân dung Khách hàng mục tiêu (Target Customer Profile):Đặt hai nhóm lên bàn cân, bạn sẽ thấy nhóm Đồng ý (Response=1) mang những đặc điểm "vàng" sau:
#-💰 Thu nhập vượt trội (Income): Nhóm đồng ý mua có thu nhập trung bình lên tới $60,076, cao hơn hẳn nhóm từ chối (~$50,244). Điều này chứng tỏ chiến dịch quảng cáo của siêu thị đang thu hút đúng tệp người có điều kiện kinh tế khá giả.
#-🔥 Sức mua cực khủng (Total_Spent): Đây là con số ấn tượng nhất! Khách hàng chịu phản hồi chiến dịch có mức chi tiêu trung bình lên đến $985.7 — gần gấp đôi so với nhóm từ chối ($540.2).
#->Insight: Nhóm phản hồi chiến dịch chính là tệp Khách hàng VIP của siêu thị. Họ không chỉ đồng ý mua sản phẩm quảng cáo mà bản thân họ đã là những người chi tiền nhiều nhất cho hệ thống.
#-⏳ Độ trung thành rất cao (Customer_Tenure): Nhóm đồng ý mua có thời gian gắn bó trung bình là 448 ngày (khoảng 1 năm 3 tháng), dài hơn nhóm từ chối (337 ngày). Khách hàng càng quen thuộc với thương hiệu thì càng dễ mở lòng với các chương trình quảng cáo mới.
#-🎂 Tuổi tác không phải là biến số quyết định (Age): Cả hai nhóm đều có độ tuổi trung bình xấp xỉ nhau (~54 - 55 tuổi). Điều này tái khẳng định tệp khách hàng cốt lõi của siêu thị này nói chung nằm ở phân khúc trung niên, nhưng để họ bấm "Đồng ý" mua hàng thì phụ thuộc vào Túi tiền (Income) và Độ thân thiết (Tenure) chứ không phải do tuổi tác.

# =====================================================================
# BƯỚC 10: TRỰC QUAN HÓA DỮ LIỆU (DATA VISUALIZATION)
# =====================================================================
import matplotlib.pyplot as plt
import seaborn as sns

# Thiết lập giao diện biểu đồ cho đẹp và chuyên nghiệp
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# --- BIỂU ĐỒ 1: Đóng góp doanh thu theo nhóm tuổi ---
# Sắp xếp lại dữ liệu từ bảng age_analysis bạn đã tính ở bước trước
age_revenue = df_clean.groupby('Age_Group', observed=False)['Total_Spent'].sum().reset_index()

sns.barplot(
    data=age_revenue, 
    x='Age_Group', 
    y='Total_Spent', 
    palette='Blues_r', 
    ax=axes[0]
)
axes[0].set_title('Tổng Chi Tiêu (Doanh Thu) Theo Nhóm Tuổi', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Nhóm Tuổi', fontsize=12)
axes[0].set_ylabel('Tổng Chi Tiêu ($)', fontsize=12)

# --- BIỂU ĐỒ 2: So sánh sức mua giữa nhóm Từ chối vs Đồng ý chiến dịch ---
sns.boxplot(
    data=df_clean, 
    x='Response', 
    y='Total_Spent', 
    palette='Set2', 
    ax=axes[1]
)
axes[1].set_title('So Sánh Mức Chi Tiêu Giữa Các Nhóm Phản Hồi', fontsize=14, fontweight='bold')
axes[1].set_xticklabels(['Từ Chối (Response=0)', 'Đồng Ý (Response=1)'])
axes[1].set_xlabel('Trạng Thái Phản Hồi Campaign', fontsize=12)
axes[1].set_ylabel('Tổng Chi Tiêu ($)', fontsize=12)

# Hiển thị cặp biểu đồ
plt.tight_layout()
plt.show()
```

###  IV. CHIẾN LƯỢC KINH DOANH ĐỘT PHÁ TỪ SỐ LIỆU ĐỊNH LƯỢNG (BUSINESS INSIGHTS)

Hệ thống frontend của Website bán hàng cần được tối ưu cấu trúc: tăng kích thước font chữ (đạt chuẩn AAA về độ tương phản), tinh giản luồng User Flow (giảm thiểu các bước click-to-buy) để hạ thấp rào cản tương tác công nghệ.

#### 2. Chiến lược Đa kênh kết hợp dựa trên Hiệu năng Transaction (Omnichannel Strategy)
* **Bằng chứng Thống kê:** Kênh Cửa hàng trực tiếp (`Store`) giữ vị thế cốt lõi với $46.2\%$ tổng đơn hàng ($12,844$ transactions). Tuy nhiên, kênh `Website` đã bám đuổi phân khúc với **$32.5\%$ thị phần đơn hàng** ($9,043$ transactions).
* **Hành động Kỹ thuật & Thương mại:**
    * **Triển khai Mô hình Tích hợp dữ liệu trực tuyến (Click-and-Collect):** Số liệu chứng minh kênh Online đang tăng trưởng mạnh nhưng hành vi gốc của tệp lớn tuổi vẫn gắn liền với thực địa. Siêu thị cần triển khai tính năng đặt hàng trực tuyến và nhận hàng tại Store. Hệ thống backend sẽ tận dụng thời điểm khách đến nhận hàng để kích hoạt thuật toán Cross-selling (Bán chéo) tại quầy.
    * **Kiến trúc Hệ thống Gợi ý (Recommendation System):** Tận dụng cấu trúc dữ liệu Online của $32.5\%$ đơn hàng để xây dựng bộ lọc cộng tác (Collaborative Filtering) trên Website, tự động cá nhân hóa rổ hàng hiển thị theo lịch sử chi tiêu của từng Account.

#### 3. Mô hình Lọc tệp đích tối ưu hóa chỉ số ROI của Campaign
* **Bằng chứng Thống kê:** Phân khúc chuyển đổi (`Response=1`) sở hữu các chỉ số phân phối vượt trội so với nhóm từ chối (`Response=0`): Thu nhập trung bình cao hơn **$+19.5\%$** ($\$60,076.7$ vs $\$50,244.3$), Sức mua trung bình bứt phá **$+82.5\%$** ($\$985.7$ vs $\$540.2$), và thời gian gắn bó hệ thống dài hơn **$+33.0\%$** ($448.1$ ngày vs $337.0$ ngày).
* **Hành động Kỹ thuật & Thương mại:**
